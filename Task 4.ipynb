### Importing Libraries

```python
from sklearn import datasets
import matplotlib.pyplot as plt 
```

### Importing Datasets

```python
iris = datasets.load_iris()
x = iris.data
y = iris.target
```

### Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
mdl = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
mdl.fit(x, y)
```

### Sklearn tree

```python
from sklearn import tree
rep_1 = tree.export_text(mdl)
print(rep_1)
```

### Decision Tree plot

```python
fig = plt.figure(figsize=(25,20))
graph = tree.plot_tree(mdl, feature_names=iris.feature_names, class_names=iris.target_names, filled=True)
```
